### Relevância do Atributo
Um pensamento a se considerar é se um (ou mais) das seis categorias de produto são relevantes para entender a compra do cliente. Ou seja, é possível determinar se o cliente que comprou uma quantidade X de uma categoria de produto vai necessariamente comprar uma quantidade Y de outra categoria? Para isto iremos treinar uma aprendizagem não supervisionada de regressão em um conjunto de dados com um atributo removido e então pontuar quão bem o modelo pode prever o atributo removido.

```python
# TODO: Fazer uma cópia do DataFrame utilizando a função 'drop' para soltar o atributo dado

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

drop = 'Detergents_Paper'
new_data = data.drop(drop ,axis=1)
label = data[drop]

# TODO: Dividir os dados em conjuntos de treinamento e teste utilizando o atributo dado como o alvo
X_train, X_test, y_train, y_test = train_test_split(new_data, label , test_size=0.25, random_state=50)

# TODO: Criar um árvore de decisão regressora e ajustá-la ao conjunto de treinamento
regressor = DecisionTreeRegressor(random_state=50)
regressor.fit(X_train, y_train)

# TODO: Reportar a pontuação da previsão utilizando o conjunto de teste
score = regressor.score(X_test, y_test)
print ("Score:", score)
```

## Pré-processamento de Dados
Nesta seção, iremos pré-processar os dados para criar uma melhor representação dos clientes ao executar um escalonamento dos dados, além de detectar os valores discrepantes.

### Escalonando Atributos
Quando os dados não são distribuídos normalmente, especialmente se a média e a mediana variam significativamente (indicando um grande desvio), é quase sempre apropriado aplicar um escalonamento não linear. Para isto utilizaremos o teste BoxCox que reduz o desvio calculando o melhor poder de transformação dos dados. Uma abordagem simplificada que pode funcionar na maioria dos casos seria aplicar o algoritmo natural.

```python
# Escalona os dados utilizando o algoritmo natural
log_data = np.log(data)

# Escalona a amostra de dados utilizando o algoritmo natural
log_samples = np.log(samples)

# Produz uma matriz de dispersão para cada par de atributos novos-transformados
pd.plotting.scatter_matrix(log_data, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
```

```python
# Mostre a amostra dados log-transformada
display(log_samples)
```

### Detecção de valores atípicos (_Outlier_)
Identificar dados discrepantes é extremamente importante no passo de pré-processamento de dados de qualquer análise porque a presença de discrepantes podem enviesar os resultados.

```python
# Encontra os pontos de dados com os maiores e menores valores para cada atributo
for feature in log_data.keys():
    
    # Calcula Q1 (25º percentil dos dados) para o atributo 
    Q1 = np.percentile(log_data[feature], 25)
    
    # Calcula Q3 (75º percentil dos dados) para o atributo 
    Q3 = np.percentile(log_data[feature], 75)
    
    # Utiliza a amplitude interquartil para calcular o passo do discrepante (1,5 vezes a variação interquartil)
    step = (Q3 - Q1)*1.5
    
    # Mostra os dados discrepantes
    print ("Data points considered outliers for the feature '{}':".format(feature))
    display(log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))])
    
# Seleciona os índices dos pontos de dados a remover
outliers  = [65, 66, 75, 128, 154]

# Remove os valores atípicos
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```

Podemos ver que há dados discrepantes principalmente nas features que tinham uma baixa correlação. Características como "Grocery" e "Fresh" que possuem uma dispersão menor têm também menos pontos de dispersão.

Antes de remover um outlier é necessário entender a sua origem e o porquê dele estar ali. Se dados importantes forem retirados erroneamente a análise pode ser enviesada. Por isto, sem outro tipo de informação em mãos, vamos retirar apenas os outliers que aparecem em mais de uma categoria (65, 66, 75, 128 e 154) para que favoreça as predições do modelo sem realizar grandes mudanças.

## Transformação de Atributo
Agora iremos utilizar a análise de componentes principais (PCA) para elaborar conclusões sobre a estrutura subjacente de dados de clientes do atacado. Dado que ao utilizar a PCA em conjunto de dados calcula as dimensões que melhor maximizam a variância, nós iremos encontrar quais combinações de componentes de atributos melhor descrevem os consumidores.

### Implementação: PCA

Agora que os dados foram escalonados em uma distribuição normal e os discrepantes necessários foram removidos, vamos aplicar a PCA para descobrir qual dimensão dos dados melhor maximizam a variância dos atributos envolvidos. Além de descobrir essas dimensões, a PCA também irá reportar a *razão da variância explicada* de cada dimensão – quanta variância dentro dos dados é explicada pela dimensão sozinha.

```python
from sklearn.decomposition import PCA

# Aplica a PCA ao ajustar os bons dados com o mesmo número de dimensões como atributos
pca = PCA(n_components=6)
pca.fit(good_data)

# Transforma a amostra de data-log utilizando o ajuste da PCA acima
pca_samples = pca.transform(log_samples)

# Gera o plot dos resultados da PCA
pca_results = vs.pca_results(good_data, pca)
```

Além disso, se uma quantidade significativa de variância é explicada por apenas duas ou três dimensões, os dados reduzidos podem ser visualizados depois.

### Implementação: Criando Clusters
Dependendo do problema, o número de clusters que você espera que estejam nos dados podem já ser conhecidos. Quando um número de clusters não é conhecido *a priori*, não há garantia que um dado número de clusters melhor segmenta os dados, já que não é claro quais estruturas existem nos dados – se existem. Entretanto, podemos quantificar a "eficiência" de um clustering ao calcular o *coeficiente de silhueta* de cada ponto de dados. O [coeficiente de silhueta](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html) para um ponto de dado mede quão similar ele é do seu cluster atribuído, de -1 (não similar) a 1 (similar). Calcular a *média* do coeficiente de silhueta fornece um método de pontuação simples de um dado clustering.

No bloco de código abaixo, você vai precisar implementar o seguinte:
 - Ajustar um algoritmo de clustering para o `reduced_data` e atribui-lo ao `clusterer`.
 - Prever o cluster para cada ponto de dado no `reduced_data` utilizando o `clusterer.predict` e atribuindo eles ao `preds`.
 - Encontrar os centros do cluster utilizando o atributo respectivo do algoritmo e atribuindo eles ao `centers`.
 - Prever o cluster para cada amostra de pontos de dado no `pca_samples` e atribuindo eles ao `sample_preds`.
 - Importar sklearn.metrics.silhouette_score e calcular o coeficiente de silhueta do `reduced_data` contra o do `preds`.
   - Atribuir o coeficiente de silhueta para o `score` e imprimir o resultado.

```python
from sklearn.mixture import GaussianMixture as GMM
from sklearn.metrics import silhouette_score

# TODO: Aplique o algoritmo de clustering de sua escolha aos dados reduzidos 
clusterer = GMM(n_components=2).fit(reduced_data)

# TODO: Preveja o cluster para cada ponto de dado
preds = clusterer.predict(reduced_data)

# TODO: Ache os centros do cluster
centers = clusterer.means_

# TODO: Preveja o cluster para cada amostra de pontos de dado transformados
sample_preds = clusterer.predict(pca_samples)

# TODO: Calcule a média do coeficiente de silhueta para o número de clusters escolhidos
score = silhouette_score(reduced_data,preds)

print (score)
```

### Implementação: Recuperação de Dados
Cada cluster apresentado na visualização acima tem um ponto central. Esses centros (ou médias) não são especificamente pontos de dados não específicos dos dados, em vez disso, são *as médias* de todos os pontos estimados em seus respectivos clusters. Para o problema de criar segmentações de clientes, o ponto central do cluster corresponde *a média dos clientes daquele segmento*. Já que os dados foram atualmente reduzidos em dimensões e escalas por um algoritmo, nós podemos recuperar a despesa representativa do cliente desses pontos de dados ao aplicar transformações inversas.

No bloco de código abaixo, você vai precisar implementar o seguinte:
 - Aplicar a transformação inversa para o `centers` utilizando o `pca.inverse_transform`, e atribuir novos centros para o `log_centers`.
 - Aplicar a função inversa do `np.log` para o `log_centers` utilizando `np.exp`, e atribuir os verdadeiros centros para o `true_centers`.


```python
# TODO: Transforme inversamento os centros
log_centers = pca.inverse_transform(centers)

# TODO: Exponencie os centros
true_centers = np.exp(log_centers)

# Mostre os verdadeiros centros
segments = ['Segment {}'.format(i) for i in range(0,len(centers))]
true_centers = pd.DataFrame(np.round(true_centers), columns = data.keys())
true_centers.index = segments
display(true_centers)
```
